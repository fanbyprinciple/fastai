```
from fastai.vision.all import *
# used to import all neccessary its of your program

path = untar_data(URLs.PETS)/'images'
# returns a path object
```

```
import random
def is_cat(x):
  if(random.random()< 0.02):
    print(x) 
  return x[0].isupper()
# labels the cats based on filename rule by the dataset creators
# cats start with uppercase


dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42, label_func=is_cat, item_tfms=Resize(224)
)
# we choose 224 because of historical reason 
# item_tfms is applied over each item
# while batch_tfms is applied over a batch
```

```
learn = cnn_learner(dls, resnet34,metric=error_rate)
# error rate is 1- accuracy 
# it automatically removes the final layer

learn.fine_tune(2)
# parameters of a pretrained model are updated by training for additional epochs
```

```
from ipywidgets import widgets

uploader = widgets.FileUpload()
uploader
```

```
img = PILImage.create(uploader.data[0])
is_cat,_, probs = learn.predict(img)
print("Is this a cat ? ",{is_cat} )
print("probably ", probs, ", I think.")
img.to_thumb(192)

```
